```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
sns.barplot(x = 'sex', y = 'survived', data = df)
```

```python
sns.barplot(x = 'sex', y = 'survived', data = df, color='lightgreen')
```

```python
sns.barplot(x = 'sex', y = 'survived', data = df, color=['lightgreen','darkgreen'], errorbar= None)
```

```python
sns.barplot(x = 'sex', y = 'survived', data = df, palette=['lightgreen','darkgreen'], errorbar= None)
```

```python
sns.barplot(x = 'sex', y = 'survived', data = df, palette=['lightgreen','darkgreen'],hue = 'sex', errorbar= None)
```

```python
sns.barplot(x = 'sex', y = 'survived', data = df, palette=['lightgreen','darkgreen'],hue = 'sex', legend = None, errorbar= None)
```

```python
sns.barplot(x = 'sex', y = 'survived', data = df, palette='hls',hue = 'sex', legend = None, errorbar= None)
```

```python
sns.barplot(y = 'sex', x = 'survived', data = df, palette='hls',hue = 'sex', legend = None, errorbar= None)
```

```python
sns.barplot(x = 'class', y = 'survived', data = df, palette='hls',hue = 'embark_town', errorbar= None)
```

```python
sns.barplot(x = 'class', y = 'survived',hue = 'embark_town',order=['Third', 'Second', 'First'] ,data = df)
```
