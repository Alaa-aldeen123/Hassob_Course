```python
import pandas as pd

df = pd.DataFrame({ 'city':['Ibb','Sanaa','Aden'],
                   'day1':[12,34,12],
                   'day2':[32,54,76],
                   'day3':[34,24,56],
                   'day4':[12,45,65],
                   'day5':[12,54,65]
})
df
```

```python
df_melt = df.melt(id_vars='city')
df_melt
```

```python
df_melt = df.melt(id_vars='city', var_name='Day', value_name='Temperture')
df_melt
```

```python
df_melt = df.melt(id_vars='city', var_name='Day', value_name='Temperture').sort_values(by='city')
df_melt
```

```python
df_melt = df.melt(id_vars='city', var_name='Day', value_name='Temperture').sort_values(by='city').reset_index(drop=True)
df_melt
```
