```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('car_malfunctions.csv')
df.info()
```

```python
y1 = df.Age
y2 = df.malfunctions
y3 = df.DaysOff

plt.subplot(2,2,1)
plt.title('Age per Malfunctions', fontsize= 14)
plt.xlabel('Age', fontsize= 12)
plt.ylabel('Malfunctions', fontsize= 12)
plt.plot(y1,y2, marker= 'o', markerfacecolor= 'white')

plt.subplot(2,2,2)
plt.title('Malfunctions per Age', fontsize= 14)
plt.xlabel('Malfunctions', fontsize= 12)
plt.ylabel('Age', fontsize= 12)
plt.plot(y2,y1, marker= 'o', markerfacecolor= 'white')

plt.subplot(2,2,3)
plt.title('Malfunctions per Days Off', fontsize= 14)
plt.xlabel('Malfunctions', fontsize= 12)
plt.ylabel('Days Off', fontsize= 12)
plt.plot(y2,y3, marker= 'o', markerfacecolor= 'white')

plt.show()
```
