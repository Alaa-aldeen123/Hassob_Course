```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('dots')
df
```

```python
sns.countplot(x = 'choice', data = df)
plt.show()
```

```python
sns.countplot(x = 'choice', hue = 'coherence', data = df)
plt.show()
```

```python
plt.title('The Count of Chices for each Chherence', fontsize = 14)
plt.xlabel('Choice', fontsize = 12)
plt.ylabel('Count', fontsize = 12)
sns.countplot(x = 'choice', hue = 'coherence', data = df)
plt.show()
```

```python
df2 = sns.load_dataset('tips')
df2
```

```python
sns.countplot(x = 'sex', data = df2)
plt.show()
```

```python
sns.countplot(x = 'sex', data = df2 , palette= 'Set2' )
plt.show()
```

```python
sns.countplot(x = 'sex', hue= 'smoker', data = df2 , palette= 'Set2' )
plt.show()
```

```python
sns.countplot(x = 'day',  data = df2  )
plt.show()
```

```python
sns.countplot(x = 'day',  data = df2, hue= 'day', palette= 'hls' )
plt.show()
```

```python
sns.countplot(x = 'day',  data = df2, hue= 'day', palette= 'hls' , order= ['Sat', 'Sun', 'Thur', 'Fri'])
plt.show()
```

```python
sns.countplot(x = 'day',  data = df2, hue= 'day', palette= 'hls' , order= df2['day'].value_counts().index)
plt.show()
```

```python
sns.countplot(x = 'day',  data = df2, hue= 'day', palette= 'hls' , order= df2['day'].value_counts(ascending= True).index)
plt.show()
```

```python
bars = sns.countplot(x = 'sex',  data = df2, hue= 'day' )


for label in bars.containers:
    bars.bar_label(label, padding=3)
```

```python
sns.countplot(x='time',data=df2)
plt.show()
```

```python
sns.countplot(x='time',data=df2, hue='sex')
plt.show()
```

```python
sns.barplot(x='time',data=df2, hue='sex')
plt.show()
```

```python
sns.barplot(x='time', y='tip',data=df2, hue='sex')
plt.show()
```

```python
sns.barplot(x='time', y='tip',data=df2, hue='sex', errorbar= None)
plt.show()
```
