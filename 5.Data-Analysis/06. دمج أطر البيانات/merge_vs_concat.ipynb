```python
import pandas as pd

df1 = pd.DataFrame({'ID': [1, 2, 3], 
                    'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 
                    'Score': [85, 90, 75]})
```

```python
df1
```

```python
df2
```

```python
merge_df = pd.merge(df1,df2, on='ID')
merge_df
```

## Inner join

```python
merged_df = df1.merge(df2, on='ID', how='inner')  
merged_df
```

```python
df1.rename(columns={'ID':'new_ID'}, inplace= True)
df1
```

```python
merge_df = pd.merge(df1, df2, left_on='new_ID', right_on='ID')
merge_df
```

```python
concat_df = pd.concat([df1,df2])
concat_df
```

```python
concat_df = pd.concat([df1,df2],keys=['df1','df2'])
concat_df
```

```python
concat_df = pd.concat([df1,df2],keys=['df1','df2'],ignore_index=True)
concat_df
```

```python
concat_df = pd.concat([df1,df2],axis=1)
concat_df
```

```python
df1.rename(columns={'new_ID':'ID'}, inplace=True)
concat_df = pd.concat([df1,df2],axis=1)
concat_df
```
