```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a basic scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")

# Display the plot
plt.show()
```

```python
df = pd.read_csv('2016.csv')
df
```

```python
sns.scatterplot(x='Economy (GDP per Capita)', y='Happiness Score', data=df)
plt.show()
```

```python
sns.scatterplot(x='Economy (GDP per Capita)', y='Happiness Score', hue='Region', data=df)
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.scatterplot(x='Economy (GDP per Capita)', y='Happiness Score', hue='Region', data=df)
plt.show()
```

```python
grid = sns.FacetGrid(df, col='Region', hue='Region', col_wrap=5)
grid.map(sns.scatterplot, 'Economy (GDP per Capita)', 'Health (Life Expectancy)' )
plt.show()
```
