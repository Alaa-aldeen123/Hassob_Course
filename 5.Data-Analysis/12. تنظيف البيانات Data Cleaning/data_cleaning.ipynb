```python
for col in df.columns:
    if col != 'gender':
        df.fillna({col:df[col].mean()}, inplace = True)
    else:
        df.fillna({col:'Unknown'}, inplace = True)

df.info()
df.head(30)
```

```python
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        # For numeric columns, fill NaNs with the mean of the column
        df.fillna({col:df[col].mean()}, inplace = True)
    else:
        # For non-numeric columns, fill NaNs with 'unknown'
        df.fillna({col:'Unknown'}, inplace = True)

df.info()
df.head(10)
```

```python
for col in df.columns:
    df.fillna({col:df[col].mean()}, inplace = True, numeric_only= True)
df.info()
df.head(30)
```

```python
for col in df.columns:
    if col != 'gender':
        # Replace values that are out of the valid range with the column's mean
        df[col] = df[col].where(~((df[col] > 100) | (df[col] < 0)), df[col].mean())


df.head(15)
```
