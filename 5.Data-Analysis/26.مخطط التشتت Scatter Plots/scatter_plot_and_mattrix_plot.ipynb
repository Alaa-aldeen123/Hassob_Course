```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('car_malfunctions.csv')
df.info()
```

```python
age = df.Age
malfun = df.malfunctions
doff = df.DaysOff
```

```python
plt.scatter(age,malfun)
plt.scatter(age,doff)

plt.show()
```

```python
plt.subplot(1,2,1)
plt.title('Age Per Malfunctions', fontsize= 14)
plt.xlabel('Age', fontsize= 13)
plt.ylabel('Malfunctions', fontsize= 13)
plt.scatter(age,malfun)

plt.subplot(1,2,2)
plt.title('Age Per Days Off', fontsize= 14)
plt.xlabel('Age', fontsize= 13)
plt.ylabel('Days Off', fontsize= 13)
plt.scatter(age,doff)

plt.show()
```

```python
plt.scatter(age,malfun, color= 'Tomato')
plt.scatter(age,doff, c='black')

plt.show()
```

### color map

```python
colors= [50,60,70,80,90,100]

plt.scatter(age,malfun, c= colors, cmap='Blues')
plt.show()
```

```python
colors= [50,60,70,80,90,100]

plt.scatter(age,malfun, c= colors, cmap='Blues')
plt.colorbar()
plt.show()
```
