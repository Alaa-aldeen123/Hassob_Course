```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('car_malfunctions.csv')

df.info()
```

```python
y1= df.malfunctions
```

```python
plt.plot(y1)
plt.show()
```

```python
y2 = df.Age

plt.plot(y1)
plt.plot(y2)
```

```python
plt.plot(y1, marker='o')
plt.plot(y2, marker='o')
plt.show()
```

```python
plt.plot(y1, marker='o', markerfacecolor='white')
plt.plot(y2, marker='o', markerfacecolor='white')
plt.show()
```

```python
plt.plot(y1, marker='s', markerfacecolor='white')
plt.plot(y2, marker='s', markerfacecolor='white')
plt.show()
```

```python
plt.plot(y1, marker='^', markerfacecolor='white')
plt.plot(y2, marker='^', markerfacecolor='white')
plt.show()
```

```python
plt.plot(y1, marker='*', markerfacecolor='white')
plt.plot(y2, marker='*', markerfacecolor='white')
plt.show()
```

```python
y3 = df.DaysOff
plt.plot(y1, marker='o', markerfacecolor='white', label= 'Malfunctions')
plt.plot(y2, marker='o', markerfacecolor='white', label= 'Ages')
plt.plot(y3, marker='o', markerfacecolor='white', label= 'Days Off')

plt.legend()
plt.show()
```

```python
df2 = pd.read_csv('car_malfunctions_2.csv')

df2.info()
```

```python
plt.title('Malfunctions per Car', fontsize= 15)
plt.xlabel('Cars', fontsize= 14)
plt.ylabel('Malfunctions', fontsize= 14)
plt.bar(df2.Name, df2.Malfunctions)

plt.show()
```

```python
plt.title('Malfunctions per Car', fontsize= 15)
plt.xlabel('Cars', fontsize= 14)
plt.ylabel('Malfunctions', fontsize= 14)
plt.bar(df2.Name, df2.Malfunctions, width= 0.5)

plt.show()
```

```python
plt.title('Malfunctions per Car', fontsize= 15)
plt.xlabel('Cars', fontsize= 14)
plt.ylabel('Malfunctions', fontsize= 14)
plt.barh(df2.Name, df2.Malfunctions, width= 0.5)

plt.show()
```

```python
plt.title('Malfunctions per Car', fontsize= 15)
plt.xlabel('Cars', fontsize= 14)
plt.ylabel('Malfunctions', fontsize= 14)
plt.barh(df2.Name, df2.Malfunctions, height= 0.5)

plt.show()
```
