# Clean and Analys Data

```python
frame.columns = [col.lower() for col in frame]
frame
```

```python
'''Using [ ... ] creates a list. When you assign this list to frame.columns, 
pandas interprets it as a list containing a single element (the entire Index returned by .str.replace(...)),
which is why your DataFrame shows the columns as tuples like (order_id,)'''
#frame. columns = [frame.columns.str.replace(' ', '_')]

frame. columns = frame.columns.str.replace(' ', '_')
frame
```

```python
frame.info()
```

```python
frame.isnull().sum().sum() / len(frame) * 100
```

```python
frame = frame.dropna()
frame.isnull().sum()
```

```python
#frame = frame.astype({'quantity_ordered': int, 'price_each': float})
```

```python
frame['quantity_ordered'].unique()
```

```python
frame['price_each'].unique()
```

```python
frame = frame[frame['quantity_ordered'] != 'Quantity Ordered']
frame = frame[frame['price_each'] != 'Price Each']
```

```python
frame = frame.astype({'quantity_ordered': int, 'price_each': float})
frame.info()
```

```python
frame['sales'] = frame['quantity_ordered'] * frame['price_each']
```

```python
months_sales = frame.groupby('month')['sales'].agg(sales_sum = 'sum').sort_values(by= 'sales_sum', ascending= False).reset_index()
months_sales
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax = sns.barplot(x= 'sales_sum', y= 'month',  data= months_sales)
plt.title('Sales Per Month')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
df = frame
df['city'] = df['purchase_address'].str.split(',').str[1].str.strip()
df
```

```python
city_quantity_sum = df.groupby('city')['quantity_ordered'].agg(quantity_sum = 'sum').sort_values(by= 'quantity_sum', ascending= False).reset_index()
city_quantity_sum
```

```python
df.city.unique()
```

```python
city_quantity_count = df.groupby('city')['quantity_ordered'].agg(quantity_count = 'count').sort_values(by= 'quantity_count', ascending= False).reset_index()
city_quantity_count
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 6))

ax1 = sns.barplot(x= 'quantity_sum', y= 'city',  data= city_quantity_sum, ax= ax1)
plt.title('Quantity_Sum Per City')
ax1.bar_label(ax1.containers[0])

ax2 = sns.barplot(x= 'quantity_sum', y= 'city',  data= city_quantity_sum, ax= ax2)
plt.title('Quantity_Count Per City')
ax2.bar_label(ax2.containers[0])
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 6))

# Adjust the vertical space between subplots
fig.subplots_adjust(hspace=0.5)

# First chart: using the 'rocket' palette for variety
sns.barplot(x='quantity_sum', y='city', data=city_quantity_sum, ax=ax1,hue= 'city', palette="rocket")
ax1.set_title('Quantity_Sum Per City')
ax1.bar_label(ax1.containers[0])

# Second chart: using the 'cool' palette for a different look
sns.barplot(x='quantity_sum', y='city', data=city_quantity_sum, ax=ax2, hue= 'city', palette="cool")
ax2.set_title('Quantity_Count Per City')
ax2.bar_label(ax2.containers[0])

plt.show()

'''
Note:
When you apply a palette, Seaborn may create multiple BarContainer objects instead of a single one,
so using just ax.bar_label(ax.containers[0]) labels only the bars in the first container.
To label all bars, you need to iterate over all containers in the axes
'''
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 6))
fig.subplots_adjust(hspace=0.5)

# First chart: Quantity_Sum Per City
sns.barplot(x='quantity_sum', y='city', data=city_quantity_sum, ax=ax1, hue= 'city', palette="rocket")
ax1.set_title('Quantity_Sum Per City')
for container in ax1.containers:
    ax1.bar_label(container)

# Second chart: Quantity_Count Per City
sns.barplot(x='quantity_count', y='city', data=city_quantity_count, ax=ax2,hue= 'city', palette="cool")
ax2.set_title('Quantity_Count Per City')
for container in ax2.containers:
    ax2.bar_label(container)

plt.show()
```
