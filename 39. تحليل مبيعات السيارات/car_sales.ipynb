```python
milage_avg = sum(df['mileage'])/len(df['mileage'])
df['mileage_level'] = ['High mileage' if i > milage_avg else 'Low_milage' for i in df['mileage']]
df.loc[:10]
```

```python
eng = pd.value_counts(df['engtype'].values, sort= False)

labels = df['engtype'].unique()
labels.sort()
sizes = eng
eng.sort_index(inplace= True)

colors = ['orange', 'yellow', 'green','red']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow= True, startangle= 120)
plt.title("Engine Type Distribution", fontsize=16)

plt.show()
```

```python
# 1. Compute the counts as a Series, then sort by the index (alphabetical by engtype)
counts = df['engtype'].value_counts(sort=False).sort_index()
labels = counts.index
sizes  = counts.values

# 2. Optional: choose colors (must match len(labels))
colors = ['orange', 'yellow', 'green', 'red']  

# 3. Plot
plt.figure(figsize=(5, 5))
wedges, texts, autotexts = plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True,
    startangle=120,
    textprops={'fontsize': 14},      # controls label font size
    pctdistance=0.75                  # push %â€texts slightly outward
)

# 4. Increase the pct (autotext) font size separately if you like:
for txt in autotexts:
    txt.set_fontsize(12)

# 5. Make it a circle (equal aspect ratio)
plt.axis('equal')

plt.title("Engine Type Distribution", fontsize=16)
plt.show()
```
